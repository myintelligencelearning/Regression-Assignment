```python
 import pandas as pd # Import the pandas library that helps you work with data in tables and give it the alias 'pd'
```

```python
dataset = pd.read_csv("insurance.csv") # Read the CSV file named "insurance.csv" into a dataframe called 'df'
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first=True).astype(int)  #Convert categorical 'smoker' and 'Sex' column to one-hot encoding column and  drop the first category
```

```python
dataset # Show the output
```

```python
dataset.columns
```

```python
ins=dataset[['age', 'bmi', 'children','charges', 'sex_male','smoker_yes']] # Assign the input variable columns to the variable named  'inp'
```

```python
ops=dataset[["charges"]]## Assign the output  'Profit' column to the variable named  'op'
```

```python
ops
```

```python
from sklearn.model_selection import train_test_split #Import the train_test_split function from the sklearn.model_selection module
x_train,x_test,y_train,y_test=train_test_split(ins,ops,test_size=0.3,random_state=0) #Split data into 70% train and 30% test
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestRegressor #Import the SVR (Support Vector Regression) model class from scikit-learn
regressor=RandomForestRegressor(criterion='poisson', n_estimators=100, max_features='log2', random_state=0) # Create an SVR model using a polynomial kernel and regularization parameter C=1000
regressor.fit(x_train,y_train) #Train the SVR model using the training data
```

```python
y_pred=regressor.predict(x_test) # Predict the target values using the test data
```

```python
from sklearn.metrics import r2_score  # Import the r2_score function from sklearn.metrics
r_score=r2_score(y_test,y_pred) # Calculate the R-squared value
```
