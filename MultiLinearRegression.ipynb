```python
import pandas as pd # Import the pandas library that helps you work with data in tables and give it the alias 'pd'
```

```python
df = pd.read_csv("insurance.csv") # Read the CSV file named "insurance.csv" into a dataframe called 'df'
```

```python
df
```

```python
df = pd.get_dummies(df).astype(int) #Convert categorical 'smoker' and 'Sex' column to one-hot encoding column and  drop the first category
```

```python
df
```

```python
df.columns
```

```python
df = pd.get_dummies(df, drop_first=True).astype(int) #Convert categorical 'smoker' column to one-hot encoding column and  drop the first category
```

```python
df
```

```python
inp = df[['age', 'bmi', 'children',
       'charges', 'sex_male','smoker_yes']] # Assign the input variable columns to the variable named  'inp'
```

```python
inp # Show the output
```

```python
op = df[["charges"]]## Assign the output  'Profit' column to the variable named  'op'
```

```python
op
```

```python
from sklearn.model_selection import train_test_split #Import the train_test_split function from the sklearn.model_selection module
x_train,x_test,y_train,y_test=train_test_split(inp, op, test_size=0.30, random_state=0) #Split data into 70% train and 30% test
```

```python
from sklearn.linear_model import LinearRegression # import the LinearRegression class from sklearn module
regressor = LinearRegression() # Create an instance of the LinearRegression model
regressor.fit(x_train,y_train) # Train the model using the training data
```

```python
weight=regressor.coef_ # Get the coefficients
weight # Display the weights
```

```python
bias = regressor.intercept_ #Get the intercept
bias # Show the bias
```

```python
y_pred=regressor.predict(x_test) # Predict the target values using the test data
```

```python
from sklearn.metrics import r2_score # Import the r2_score function from sklearn.metrics
r_score = r2_score(y_test, y_pred)  # Calculate the R-squared value
r_score  # Show the R-squared score
```

```python
import pickle # Import the pickle module for saving and loading models
filename ="final_model_multilinear.sav"  # Name of the file to save the model
pickle.dump(regressor, open(filename,'wb')) # Save the model to the file
loaded_regressor = pickle.load(open("final_model_multilinear.sav","rb"))  # Load the model from the file
```
